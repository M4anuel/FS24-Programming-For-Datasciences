```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use("ggplot")
```

```python
data = np.random.multivariate_normal([0,0], [[5, 2],[2,2]], size=1000)
data = pd.DataFrame(data, columns=["x", "y"])

with sns.axes_style("white"):
    sns.jointplot(data["x"], kind="kde")
plt.show()

plt.subplot(1,2,1)
for col in "xy":
    plt.hist(data[col], density=True, alpha=0.5)
plt.subplot(1,2,2)
sns.kdeplot(data, fill=True)
plt.show()
```

```python
x = np.linspace(0,10,50)
dy = 0.1
y = np.sin(x) + (dy * np.random.randn(50))
plt.axes([0,0,1,0.9])
plt.plot(x,np.sin(x))
ax2=plt.axes([1.1, 0.5, 0.5, 0.4])
plt.plot(x,np.sin(x))
plt.errorbar(x,y, yerr=dy, fmt=".k")#, capsize=100, elinewidth=20, ecolor="blue")
ax2=plt.axes([1.1, 0, 0.5, 0.4])
plt.plot(x,np.sin(x))
plt.fill_between(x, y+1, y-1, alpha=0.3)
plt.plot()

plt.show()
```
